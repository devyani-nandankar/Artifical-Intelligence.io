```python
from itertools import permutations

def solve_cryptarithmetic():
    letters = ('S', 'E', 'N', 'D', 'M', 'O', 'R', 'Y')
    digits = range(10)

    solutions = []

    # Try all possible assignments
    for perm in permutations(digits, len(letters)):
        assign = dict(zip(letters, perm))

        # Constraint 1: Leading digits cannot be zero
        if assign['S'] == 0 or assign['M'] == 0:
            continue

        # Form numbers
        SEND = (assign['S'] * 1000 +
                assign['E'] * 100 +
                assign['N'] * 10 +
                assign['D'])

        MORE = (assign['M'] * 1000 +
                assign['O'] * 100 +
                assign['R'] * 10 +
                assign['E'])

        MONEY = (assign['M'] * 10000 +
                 assign['O'] * 1000 +
                 assign['N'] * 100 +
                 assign['E'] * 10 +
                 assign['Y'])

        # Constraint 2: Equation must satisfy
        if SEND + MORE == MONEY:
            solutions.append(assign)

            print("\nSolution Found:")
            print(assign)
            print(f"{SEND} + {MORE} = {MONEY}")

    return solutions


# Run the solver
solve_cryptarithmetic()
```
