Write a Python Program to implement magic square.

```python
def generate_magic_square(n):
    magic_square = [[0 for x in range(n)] for y in range(n)]

    num = 1
    i = 0
    j = n // 2

    while num <= n * n:
        magic_square[i][j] = num
        num += 1

        new_i = (i - 1) % n
        new_j = (j + 1) % n

        if magic_square[new_i][new_j]:
            i = (i + 1) % n
        else:
            i = new_i
            j = new_j

    return magic_square


# User Input
n = int(input("Enter an odd number: "))

if n % 2 == 0:
    print("Magic square works only for odd numbers.")
else:
    square = generate_magic_square(n)

    print("\nMagic Square:")
    for row in square:
        for value in row:
            print(value, end="\t")
        print()
```
