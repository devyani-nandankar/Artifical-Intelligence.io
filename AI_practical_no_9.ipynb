```python
# Bayes' Rule for Rain Prediction

# Given probabilities
P_rain = 0.3
P_no_rain = 1 - P_rain

P_cloudy_given_rain = 0.8
P_cloudy_given_no_rain = 0.4

# Step 1: Calculate total probability of evidence (Cloudy)
P_cloudy = (P_cloudy_given_rain * P_rain) + \
           (P_cloudy_given_no_rain * P_no_rain)

# Step 2: Apply Bayes' Theorem
P_rain_given_cloudy = (P_cloudy_given_rain * P_rain) / P_cloudy

# Output
print("P(Cloudy) =", P_cloudy)
print("P(Rain | Cloudy) =", P_rain_given_cloudy)
```
