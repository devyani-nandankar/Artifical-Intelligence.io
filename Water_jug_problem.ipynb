```python
from collections import deque

# capacities of jugs
jug4 = 4
jug3 = 3

# goal
goal = 2

# visited states
visited = set()

# queue for BFS
queue = deque()

# initial state
queue.append((0, 0, []))

while queue:
    x, y, path = queue.popleft()

    if (x, y) in visited:
        continue

    visited.add((x, y))

    path = path + [(x, y)]

    # goal condition
    if x == goal:
        print("Solution Steps:")
        for step in path:
            print(step)
        break

    next_states = []

    # Rule 1: Fill 4 gallon jug
    next_states.append((jug4, y))

    # Rule 2: Fill 3 gallon jug
    next_states.append((x, jug3))

    # Rule 3: Empty 4 gallon jug
    next_states.append((0, y))

    # Rule 4: Empty 3 gallon jug
    next_states.append((x, 0))

    # Rule 5: Pour 4 -> 3
    transfer = min(x, jug3 - y)
    next_states.append((x - transfer, y + transfer))

    # Rule 6: Pour 3 -> 4
    transfer = min(y, jug4 - x)
    next_states.append((x + transfer, y - transfer))

    for state in next_states:
        queue.append((state[0], state[1], path))
```
