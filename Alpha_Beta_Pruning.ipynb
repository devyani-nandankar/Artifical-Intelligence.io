```python
import math

# Alpha-Beta Pruning Function
def alpha_beta(depth, node_index, is_max, values, alpha, beta):

    # Base case: leaf node
    if depth == 2:
        return values[node_index]

    if is_max:
        best = -math.inf

        for i in range(2):
            val = alpha_beta(depth + 1, node_index * 2 + i,
                             False, values, alpha, beta)
            best = max(best, val)
            alpha = max(alpha, best)

            # Pruning condition
            if beta <= alpha:
                break

        return best

    else:
        best = math.inf

        for i in range(2):
            val = alpha_beta(depth + 1, node_index * 2 + i,
                             True, values, alpha, beta)
            best = min(best, val)
            beta = min(beta, best)

            # Pruning condition
            if beta <= alpha:
                break

        return best


# -------------------------
# DRIVER CODE
# -------------------------
values = [3, 5, 6, 9]  # Leaf nodes

result = alpha_beta(
    depth=0,
    node_index=0,
    is_max=True,
    values=values,
    alpha=-math.inf,
    beta=math.inf
)

print("Optimal Value:", result)
```
